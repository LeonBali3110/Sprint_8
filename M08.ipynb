### PCA

El anÃ¡lisis de componentes principales (PCA) es principalmente un algoritmo de reducciÃ³n de dimensionalidad. Esto sirve para cuando tenemos un dataset muy grande, ahorar tiempo en el modelado. El PCA resume nuestras columnas para que la computadora lo entienda mÃ¡s rapido, lo ideal es no usarlo, ya que se puede perder extitud, pero a veces esto es necesario.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA().fit(df_1)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca = PCA(1)  
projected = pca.fit_transform(df_1)
print(df_1.shape)
print(projected.shape)
```

Primero cree una grafica para ayudarme a escoger las dimenciones mÃ¡s adecuadas para la reducciÃ³n. por supuesto, queremos reducirlo lo mÃ¡s posible sin sacrificar informaciÃ³n en el proceso, y es por eso que escogi 1, pues llega a 0.95, que esta muy cerca de 1.00 y es la maxima reducciÃ³n

```python
Dummies = [D_Pub, D_Sem, D_Hora]
```

```python
df_pca = pd.DataFrame(projected, columns=['PCA'])
```

### Modelos

Los modelos que utilizaremos seran K-means y Agglomerative. Pero para eso primero necesitamos identidifar cuantos clusters vamos a usar en cada caso.

```python
 from sklearn.cluster import AgglomerativeClustering
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_pca, method='ward'))
```

En dendrograma separa de lo general a lo concreto, y eso es lo que vemos, al progroma separandolo en clusters, adicinalmente los separa en colores para recomendarnos en cuantos clusters deberiamos separarlos, y en este caso nos recomienda separarlos en 3. Tambien, existe una teoria que dice, que la opciÃ³n mÃ¡s optima es, la suma de los puntos de la linea divisiva (imaginaria) que atraviesa la linea mÃ¡s larga de la dos primeras, es decir, 3. Por lo que usaremos 3 clusters, para AgglomerativeClustering.

Vamos a utilizar K-means y AgglomerativeClustering. Y para eso necesitaremos medir los clusters ideales para cada uno.

Concatenar: projected y dummies estan separados, como juntarlos o si se deben juntar.

```python
from sklearn.cluster import KMeans
```

```python
kmeans_kwargs = {
     "init": "random",
     "n_init": 10,
     "max_iter": 300,
     "random_state": 42,
}

sse = []
for k in range(1, 11):
    kmeans_c = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans_c.fit(df_pca)
    sse.append(kmeans_c.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

El MÃ©todo del codo se hace ejecutan varios k -means, incremÃ©ntanlos k con cada iteraciÃ³n y registramos el SSE.

Hay un punto Ã³ptimo donde la curva SSE comienza a doblarse, conocido como punto del codo.

Esta grÃ¡fica nos ayuda a identificar el nÃºmero de clusters que utilizaremos en nuestro modelo, en este caso (Metodo del Codo) sera para K-means. El quiebre se da en 2 pero, el codo y aquel que es el mÃ¡s optimo esta entre 3-4, escogere 4 para este caso.

```python
clustering = AgglomerativeClustering(n_clusters = 3).fit(df_pca)
```

```python
clustering.labels_
```

```python
clus_ar = clustering.labels_
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 0).fit(df_pca)
```

```python
kmeans.labels_
```

```python
Karray = kmeans.labels_
```
